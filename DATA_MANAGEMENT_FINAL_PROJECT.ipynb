```python
%%sql
DROP TABLE IF EXISTS SUB Cascade;

CREATE TABLE SUB (
    adsh VARCHAR(20) NOT NULL,
    cik NUMERIC(10) NOT NULL,
    name VARCHAR(150) NOT NULL,
    sic NUMERIC(4),
    countryba VARCHAR(2),
    stprba VARCHAR(2),
    cityba VARCHAR(30),
    zipba VARCHAR(10),
    bas1 VARCHAR(40),
    bas2 VARCHAR(40),
    baph VARCHAR(20),
    countryma VARCHAR(2),
    stprma VARCHAR(2),
    cityma VARCHAR(30),
    zipma VARCHAR(10),
    mas1 VARCHAR(40),
    mas2 VARCHAR(40),
    countryinc VARCHAR(3),
    stprinc VARCHAR(2),
    ein NUMERIC(10),
    former VARCHAR(150),
    changed DATE,
    afs VARCHAR(5),
    wksi BOOLEAN,
    fye VARCHAR(4),
    form VARCHAR(10),
    period DATE,
    fy NUMERIC(4),
    fp VARCHAR(2),
    filed DATE,
    accepted DATE,
    prevrpt BOOLEAN,
    detail BOOLEAN,
    instance VARCHAR(32),
    nciks NUMERIC(4),
    aciks VARCHAR(120),
    PRIMARY KEY (adsh)
);
```

```python
%%sql
DROP TABLE IF EXISTS TAG Cascade;

CREATE TABLE TAG (
    tag VARCHAR(2800) NOT NULL,
    version VARCHAR(20) NOT NULL,
    custom VARCHAR(600),
    abstract VARCHAR(10),
    datatype VARCHAR(20),
    iord VARCHAR(400),
    crdr VARCHAR(10),
    tlabel VARCHAR(512),
    doc VARCHAR(20000),
    PRIMARY KEY (tag, version)
);
```

```python
%%sql
DROP TABLE IF EXISTS NUM CASCADE;

CREATE TABLE NUM (
    adsh VARCHAR(20),
    tag VARCHAR(256),
    version VARCHAR(20),
    coreg VARCHAR(256) default ' ',
    ddate DATE,
    qtrs NUMERIC(8),
    uom VARCHAR(20),
    value NUMERIC(28,4) NULL,
    footnote VARCHAR(512) NULL --,
   -- primary key (adsh,tag,version,coreg,ddate,qtrs,uom)
);
```

##### Q1)  Write a query that lists the registrant’s name, their Standard Industrial Classification (SIC), the submission type and the number of times a submission type was submitted.  Indicate only those records where the country of the registrant's business address is Great Britain

```python
%%sql
SELECT name as "Registrant Name", sic as "Standard Industrial Classification" , form as "Submission Type", COUNT(form) AS submission_count
FROM SUB
WHERE countryba = 'GB'
GROUP BY name, sic, form
order by count(form) desc
LIMIT 20;
```

##### Q2) Write a query that lists the registrant’s name, tag, version, and sum of all values per tag for end date 20230131. Only present tag records when the tag is used to represent a numeric fact (abstract <>1).

```python
%%sql
SELECT
    sub.name AS "Registrant Name",
    num.tag AS Tag,
    num.version AS Version,
    SUM(num.value) AS "Total Value"
FROM sub
 JOIN num ON sub.adsh = num.adsh
 JOIN tag ON num.tag = tag.tag AND num.version = tag.version
WHERE num.ddate = '20230131'
  AND tag.abstract <> '1'
GROUP BY sub.name, num.tag, num.version
ORDER BY SUM(num.value) desc nulls last
LIMIT 20;
```

##### Q3) Write a query that lists the registrant’s name, their business address (city and zip code only) and ein where the country of the registrant's business address or the country of incorporation for the registrant is Great Britain. Use set operation.

```python
%%sql
SELECT
    name AS "Registrant Name",
    cityba AS City,
    zipba AS "Zip Code",
    Ein

FROM SUB
WHERE
    countryba = 'GB'

UNION

SELECT
    name,
    cityba,
    zipba,
    ein
FROM SUB
WHERE
    countryinc = 'GB'
LIMIT 20;
```

##### Q4) Write a query that lists the registrant’s name, their former names (if not null or is similar to the registrant’s name), the date a name change occurred and how long compared to latest submission date.

```python
%%sql
SELECT
    s.name AS "Registrant Name",
    s.former AS "Former Name",
    s.changed  AS "Name Change Date",
    s.filed - s.changed AS "# of days since change"
FROM
    sub s
WHERE
    s.former IS NOT NULL
    AND s.former <> s.name
    LIMIT 20;
```

##### QUERY 1: Select all firms in the United States with net losses greater than 1,000,000 for the first quarter that changed their name within that time frame.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%%sql result <<
SELECT s.adsh, s.name, s.countryma, s.changed, ABS(n.value) AS abs_value
FROM SUB s
JOIN NUM n ON s.adsh = n.adsh
WHERE s.countryma = 'US' -- Firms in the United States
AND n.qtrs = 1 -- First quarter
AND n.value < -1000000 -- Net losses greater than 1,000,000
AND s.changed >= DATE_TRUNC('quarter', n.ddate) -- Changed name within the first quarter
AND s.changed < DATE_TRUNC('quarter', n.ddate) + INTERVAL '3 months'; -- Name changed within the first quarter
```

##### QUERY 2 - Write a query that lists the tag, number of quarters, whether the tag is custom or not, only for co-registrants.

```python
%%sql result_quarters <<
SELECT DISTINCT
    NUM.tag as Tag,
    COUNT (DISTINCT NUM.ddate) AS "Number of Quarters",
    CASE
        WHEN CAST(TAG.custom AS INTEGER) = 1 THEN 'Yes'
        ELSE 'No'
    END AS "Custom Tag"
FROM
    SUB
JOIN
    NUM ON SUB.adsh = NUM.adsh
JOIN
    TAG ON NUM.tag = TAG.tag AND NUM.version = TAG.version
WHERE
    NUM.coreg IS NOT NULL
    AND TAG.abstract <> '1'
GROUP BY
    NUM.tag, TAG.custom
ORDER BY
    NUM.tag;
```

```python
df = pd.DataFrame(result_quarters)
df.head(1)

counts_dict = {}

for value in df['Number of Quarters']:
    if value in counts_dict:
        counts_dict[value] += 1
    else:
        counts_dict[value] = 1
```

```python
pd.set_option('display.max_columns',None)
count_df = pd.DataFrame(counts_dict, index = counts_dict.keys())

df_simple = pd.DataFrame(count_df.iloc[0], index=counts_dict.keys())
```

```python
larger_than_threshold = df_simple[df_simple[[1]] > 400]
smaller_than_threshold = df_simple[df_simple[[1]] <= 400]

larger_than_threshold.dropna(inplace=True)
```

##### QUERY 3 - REPORT ANY FIRMS THAT FILED A 14A DOCUMENT AND REPORT THE NET INCOME OR LOSS FOR BOTH DOMESTIC AND FOREIGN FIRMS THAT OPERATED WITH A LARGE ACCELERATED SUBMISSION TIMEFRAME

```python
%%sql foreign14A_positive <<
SELECT DISTINCT s.adsh, s.name, s.countryba, s.form, n.tag, n.value, n.uom
FROM SUB s
JOIN NUM n ON s.adsh = n.adsh
WHERE s.form IN ('DEF 14A', 'PRE 14A') --filed a 14A
    AND s.countryba != 'US' --foreign firms
    AND s.afs = '1-LAF' --large accelerated timeframe
    AND n.value > 300000000
    AND n.tag != 'NetIncomeLoss'; --PROFIT
```

```python
%%sql foreign14A_loss <<
SELECT DISTINCT s.adsh, s.name, s.countryba, s.form, n.tag, n.value, n.uom
FROM SUB s
JOIN NUM n ON s.adsh = n.adsh
WHERE s.form IN ('DEF 14A', 'PRE 14A')
    AND s.countryba != 'US' --foreign firms
    AND s.afs = '1-LAF'
    AND n.value > 300000000
    AND n.tag = 'NetIncomeLoss'; --LOSS
```

```python
%%sql domestic14A_positive <<
SELECT DISTINCT s.adsh, s.name, s.countryba, s.form, n.tag, n.value, n.uom
FROM SUB s
JOIN NUM n ON s.adsh = n.adsh
WHERE s.form IN ('DEF 14A', 'PRE 14A')
    AND s.countryba = 'US' --domestic firms
    AND s.afs = '1-LAF'
    AND n.value > 300000000
    AND n.tag != 'NetIncomeLoss'; --PROFIT
```

```python
%%sql domestic14A_loss <<
SELECT DISTINCT s.adsh, s.name, s.countryba, s.form, n.tag, n.value, n.uom
FROM SUB s
JOIN NUM n ON s.adsh = n.adsh
WHERE s.form IN ('DEF 14A', 'PRE 14A')
    AND s.countryba = 'US' --domestic firms
    AND s.afs = '1-LAF'
    AND n.value > 300000000
    AND n.tag = 'NetIncomeLoss'; --LOSS
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df_foreign14A_positive = pd.DataFrame(foreign14A_positive)
df_foreign14A_loss = pd.DataFrame(foreign14A_loss)
df_domestic14A_positive = pd.DataFrame(domestic14A_positive)
df_domestic14A_loss = pd.DataFrame(domestic14A_loss)
```

```python
fPOS_sum = df_foreign14A_positive['value'].sum()
fLOSS_sum = df_foreign14A_loss['value'].sum()
dPOS_sum = df_domestic14A_positive['value'].sum()
dLOSS_sum = df_domestic14A_loss['value'].sum()

fPOS_mean = df_foreign14A_positive['value'].mean()
fLOSS_mean = df_foreign14A_loss['value'].mean()
dPOS_mean = df_domestic14A_positive['value'].mean()
dLOSS_mean = df_domestic14A_loss['value'].mean()
```
